```python
import pandas as pd
import numpy as np

import re
import string
import os
import jieba
import gensim
from gensim import corpora

# libraries for visualization
import pyLDAvis
import pyLDAvis.gensim
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
output_path = 'E:\MY Documents\Desktop\int\任务12'
file_path = 'E:\MY Documents\Desktop\int\任务12'
os.chdir(file_path)
data1=pd.read_excel("projects.xlsx")
data=data1.T
data.reset_index(drop=True, inplace=True)
os.chdir(output_path)
dic_file = "E:\MY Documents\Desktop\int\任务12\words.txt"
stop_file = "E:\MY Documents\Desktop\int\任务12\stopwords.txt"
```

```python
with open(dic_file,encoding = 'utf-8') as f:
    text = f.read()
    words= text.split()
for i in words:
    jieba.add_word(i, freq=None, tag=None)
```

```python
data["content"] = data.iloc[0:]
data.content=data.content.astype(str)
print(data.content.head())
```

```python
def find_chinese(file):
    pattern = re.compile(r'[^\u4e00-\u9fa5]')
    chinese = re.sub(pattern, '', file)
    return chinese
s = data.content[2]
s1 = find_chinese(s)
print(type(s1))
```

```python
with open(stop_file,encoding = 'utf-8') as f:
    text = f.read()
    words= text.split()
    
def wor_cut(text):
    s=text
    s1 = find_chinese(s)
    cut = jieba.cut(s1)
    final =[]
    for x in cut:
        if x not in words:
            final.append(x)
    return ','.join(final)

s = data.content[0]
print(wor_cut(s))

data["content_cutted"] = data.content.apply(wor_cut)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
def print_top_words(model, feature_names, n_top_words):
    tword = []
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        topic_w = " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]])
        tword.append(topic_w)
        print(topic_w)
    return tword
```

```python
n_features = 1000 #提取1000个特征词语
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                max_features=n_features,
                                stop_words='english',
                                max_df = 0.5,
                                min_df = 10)
tf = tf_vectorizer.fit_transform(data.content_cutted)
```

```python
n_topics = 10
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=50,
                                learning_method='batch',
                                learning_offset=50,
#                                 doc_topic_prior=0.1,
#                                 topic_word_prior=0.01,
                               random_state=0)
lda.fit(tf)
```

```python
n_top_words = 25
tf_feature_names = tf_vectorizer.get_feature_names()
topic_word = print_top_words(lda, tf_feature_names, n_top_words)
```

```python
import numpy as np
topics=lda.transform(tf)
```

```python
import pyLDAvis
import pyLDAvis.sklearn
```

```python
topic = []
for t in topics:
    topic.append("Topic #"+str(list(t).index(np.max(t))))
data['概率最大的主题序号']=topic
data['每个主题对应概率']=list(topics)
print(data.head())
data.to_excel("topic_full_chinese_v1_1.xlsx", engine='xlsxwriter')
```
